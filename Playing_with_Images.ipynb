```python
import  PIL, google.colab  
import matplotlib.pyplot as plt
from PIL import ImageEnhance
```

#Original Image

```python
google.colab.files.upload()
chithram1=PIL.Image.open('picture.png')
plt.imshow(chithram1)
```

#Rotated Image

```python
for i in range(0,180,10):
  Rotate=chithram1.rotate(i)
  plt.imshow(Rotate)
```

#Blured Image

```python
chithram2=chithram1.resize((50,50))
plt.imshow(chithram2)
```

#Image Size

```python
print(chithram1.size)
```

#Image Brightness

```python
Bright = ImageEnhance.Brightness(chithram1)
Bright_chithram = Bright.enhance(2.8)
plt.imshow(Bright_chithram)
```

#Image Sharpness

```python
Sharpness = ImageEnhance.Sharpness(chithram1)
Sharpness_chithram = Sharpness.enhance(7)
plt.imshow(Sharpness_chithram)
```

#Image Contrast

```python
Contrast = ImageEnhance.Contrast(chithram1)
Contrast_chithram = Contrast.enhance(3.1)
plt.imshow(Contrast_chithram)
```

#Image Color

```python
Color = ImageEnhance.Color(chithram1)
Color_chithram = Color.enhance(3.1)
plt.imshow(Color_chithram)
```
